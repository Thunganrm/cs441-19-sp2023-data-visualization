````python

import pandas as pd

from google.colab import files
 
 
uploaded = files.upload()
uploaded = files.upload()

data1 = pd.read_csv('state_labels.csv')
data2 = pd.read_csv('test.csv')

merged_data = pd.merge(data1, data2, left_on='StateID', right_on="State")
print(merged_data)
````

````python
import plotly.express as px
fig = px.pie(merged_data, names='StateName', title='State location in Malaysia')
fig.show()
````

# The pie chart depicts the alocation state of pet adoption in Malaysia. There are a total of 15 states in Malaysia that allow pet adoption. The most common state for pet adoption is Senglagor with 46.1%.

````python
from plotly.offline import init_notebook_mode,iplot
import plotly.graph_objects as go
aver_age=[]
qty = []
for i in list(data2['Type'].unique()):
    aver_ages = data2[data2['Type']==i]['Age'].mean()
    qtys = data2[data2['Type']==i]['Quantity'].sum()
    aver_age.append(aver_ages)
    qty.append(qtys)

trace = go.Bar(x = ["Dog","Cat"],y=aver_age,marker=dict(color=qty,colorscale='Viridis',showscale=True),text = qty)
data=[trace]
layout = go.Layout(title='Average of Age and Total of quantity',xaxis=dict(title='Type of animals'),yaxis=dict(title='Average of Age'),hovermode='closest')
figure = go.Figure(data=data,layout=layout)
iplot(figure)
````

# The bar chart shows the average age and quantity of two types of dog and cat. The average age of dogs is 8.11 while that of cats is 14.13. The color of each column represents the number of each species. Specifically yellow is 3177 and purple is 3167.
