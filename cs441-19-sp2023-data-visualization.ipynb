# 2. Pclass: ticket class => **Histogram**

````python

plt.figure(figsize=(10, 8))
sns.histplot(x='Pclass', data = df, bins=50)
plt.xlabel('Class', fontsize=16)
plt.ylabel('Quanity', fontsize=16)
plt.title("Histogram of Ticker class", fontsize=18)
````

#3. Name => **devine into First name and Last name**

````python
df[["First Name","Last name"]]=df['Name'].str.split(',', expand=True)
print(df[["First Name","Last name","Name"]])
````

#4. Sex => **Histogram**

````python
plt.figure(figsize=(10, 8))
sns.histplot(x='Sex', data = df, bins=50, kde=True)
plt.xlabel('Genre', fontsize=16)
plt.ylabel('Quanity', fontsize=16)
plt.title("Histogram of Sex/Genre", fontsize=18)
````

# 5.Age => Histogram

````python
plt.figure(figsize=(10, 8))
sns.histplot(x='Age', data = df, bins=100, kde=True)
plt.xlabel('Age', fontsize=16)
plt.ylabel('Quanity', fontsize=16)
plt.title("Histogram of Passenger's Age", fontsize=18)
````

# 6.SibSp => Histogram

````python
plt.figure(figsize=(10, 8))
sns.histplot(x='SibSp', data = df, bins=16, kde=True)
plt.xlabel('Sibling/Spouse', fontsize=16)
plt.ylabel('Quanity', fontsize=16)
plt.title("Histogram of Passenger's Sibling/Spouse aboard the Titanic", fontsize=18)
````

````python
plt.figure(figsize=(15, 8))
sns.histplot(x='Parch', data = df, bins=50, kde=True)
plt.xlabel('parents / children', fontsize=16)
plt.ylabel('Quanity', fontsize=16)
plt.title("Histogram of Passenger's parents / children aboard the Titanic", fontsize=18)
````

# 7. Fare=> Histogram

````python
plt.figure(figsize=(15, 8))
sns.histplot(x='Fare', data = df, bins=50, kde=True)
plt.xlabel('Price ', fontsize=16)
plt.ylabel('Quanity', fontsize=16)
plt.title("Histogram of Ticket fare", fontsize=18)
````

````python
plt.figure(figsize=(15, 8))
sns.histplot(x='Embarked', data = df, bins=1, kde=True)
plt.xlabel('Embarked ', fontsize=16)
plt.ylabel('Quanity', fontsize=16)
plt.title("Histogram of Embarked", fontsize=18)
````
